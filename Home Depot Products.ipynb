```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.style.use('default')
plt.style.use('dark_background')
```

```python
products = pd.read_csv('smart_home_gifts.csv')

products
```

```python
products.groupby('skus.info.brandName')['skus.ratingsReviews.totalReviews'] \
    .sum() \
    .sort_values(ascending=False) \
    .head(5) \
    .plot(kind='barh')
```

```python
products.groupby('skus.info.brandName')['skus.storeSku.pricing.percentageOff'] \
    .mean() \
    .sort_values(ascending=False) \
    .plot(kind='pie')

plt.axis('off')
plt.show()
plt.close()
```
